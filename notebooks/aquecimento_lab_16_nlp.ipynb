## 🔥 Aquecimento 1: A Faxina do Texto (O Acessor `.str`)
O primeiro passo da PNL é padronizar. A máquina não entende que "BOM", "bom" e "bom!!!" são a mesma palavra. O nosso objetivo aqui é esmagar tudo para letras minúsculas e arrancar a pontuação usando o acessor `.str` do Pandas.

**Sua Tarefa:** Execute o código abaixo para gerar nossos dados simulados e aplique os métodos de limpeza.

```python
import pandas as pd

# Gerando dados simulados de avaliações de clientes
dados_texto = pd.DataFrame({
    'Cliente': ['João', 'Maria', 'Pedro', 'Ana'],
    'Review_Bruta': [
        'Odiei o produto, péssimo!!!', 
        'Celular EXCELENTE... recomendo.', 
        'NÃO COMPREM! Bateria horrível.',
        'Produto bom, mas a entrega atrasou.'
    ]
})

print("--- Dados Originais ---")
display(dados_texto)

# ESCREVA SEU CÓDIGO AQUI: 

# 1. Converta todo o texto para letras minúsculas usando .str.lower()
dados_texto['Review_Limpa'] = dados_texto['Review_Bruta'].str.lower()

# 2. Remova as pontuações usando .str.replace() com Regex
# A expressão '[^\w\s]' significa: "Substitua tudo que NÃO for palavra ou espaço por nada ('')"
dados_texto['Review_Limpa'] = dados_texto['Review_Limpa'].str.replace(r'[^\w\s]', '', regex=True)

print("\n--- Dados Após a Faxina ---")
display(dados_texto[['Cliente', 'Review_Limpa']])
```

## 🔥 Aquecimento 2: A Gramática da Máquina (Bag-of-Words)
Agora que o texto está limpo, precisamos traduzi-lo para a matemática. O Scikit-Learn possui a classe `CountVectorizer`. O que ela faz? Ela lê todas as frases, encontra cada palavra única e cria uma coluna para ela. Se a palavra aparece na frase, ela ganha `1`. Se não aparece, ganha `0`.

Além disso, vamos aproveitar para remover as **Stopwords** (palavras que não agregam valor analítico, como artigos e conectivos).

**Sua Tarefa:** Instancie o `CountVectorizer`, aplique aos textos limpos e veja a matriz matemática nascer!

```python
from sklearn.feature_extraction.text import CountVectorizer

# Lista básica de Stopwords em português para o nosso aquecimento
minhas_stopwords = ['o', 'a', 'os', 'as', 'mas', 'e', 'de', 'do', 'da']

# ESCREVA SEU CÓDIGO AQUI:

# 1. Instancie o vetorizador passando a lista de stopwords para ele ignorar
vetorizador = CountVectorizer(stop_words=minhas_stopwords)

# 2. Fase de Estudo e Prova (fit_transform): Aprenda as palavras e crie a matriz
matriz_matematica = vetorizador.fit_transform(dados_texto['Review_Limpa'])

# 3. Extraindo os nomes das colunas (as palavras que a máquina aprendeu)
palavras_aprendidas = vetorizador.get_feature_names_out()

# Convertendo a matriz que o Scikit-Learn gerou de volta para um DataFrame do Pandas para podermos visualizar
df_bag_of_words = pd.DataFrame(
    matriz_matematica.toarray(), 
    columns=palavras_aprendidas
)

print("--- A Matriz Matemática (Bag of Words) ---")
display(df_bag_of_words)
```
